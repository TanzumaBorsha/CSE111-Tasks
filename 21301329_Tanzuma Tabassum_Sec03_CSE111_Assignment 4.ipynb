```python
#Task 1
class Calculator:

    def __init__(self):
        print("Let's Calculate!")

    def add(self,first_value,second_value):
        result=first_value+second_value
        print("Result: "+str(result))

    def subtract(self,first_value,second_value):
        result=first_value-second_value
        print("Result: "+str(result))

    def multiply(self,first_value,second_value):
        result=first_value*second_value
        print("Result: "+str(result))

    def divide(self,first_value,second_value):
        if second_value==0:
            print("The second value can not be 0, since division by zero is not possible")
            return

        result=first_value/second_value
        print("Result: "+str(result))


obj = Calculator()
first_value=int(input("Value 1: "))
operator=input("Operator: ")
second_value=int(input("Value 2: "))

if operator=='+':
    obj.add(first_value,second_value)
elif operator=='-':
    obj.subtract(first_value,second_value)
elif operator=='*':
    obj.multiply(first_value,second_value)
elif operator=='/':
    obj.divide(first_value,second_value)
else:
    print("You entered a incorrect operator")
```

```python
#Task 2
class Customer:

    def __init__(self, name):
        self.name = name

    def greet(self,name=None):
        print("Hello!")

    def purchase(self,*args):
        print(str(self.name)+", you purchase "+str(len(args))+" item(s):")
        for i in args:
            print(i)


customer_1 = Customer("Sam")
customer_1.greet()
customer_1.purchase("chips", "chocolate", "orange juice")
print("---------------------------------")
customer_2 = Customer("David")
customer_2.greet()
customer_2.purchase("orange juice")
```

```python
#Task 3
class Panda:
    name=""
    gender=""
    age=0

    def __init__(self,name,gender,age):
        self.name=name
        self.gender=gender
        self.age=age

    def sleep(self,time=None):
        if time==None:
            st=self.name+"s duration is unknown thus should have only bamboo leaves"
        else:
            st=self.name+" sleeps "+str(time)+" hours daily and should have "

            if time>=3 and time<=5:
                st=st+"Mixed Veggies"
            elif(time>=6 and time <=8):
                st=st+"Eggplant & Tofu"
            elif(time>=9 and time<=11):
                st=st+"Broccoli Chicken"
        return st


panda1=Panda("Kunfu","Male",5)
panda2=Panda("Pan Pan","Female",3)
panda3=Panda("Ming Ming","Female",8)

print("{} is a {} Panda Bear who is {} years old".format(panda1.name,panda1.gender,panda1.age))
print("{} is a {} Panda Bear who is {} years old".format(panda2.name,panda2.gender,panda2.age))
print("{} is a {} Panda Bear who is {} years old".format(panda3.name,panda3.gender,panda3.age))
print("===========================")
print(panda2.sleep(10))
print(panda1.sleep(4))
print(panda3.sleep())
```

```python
#Task 5
class Student:
    total=0

    def __init__(self, name=None):
        self.name=name

    def quizcalc(self,*num):
        for i in num:
            self.total+=i

    def printdetail(self):
        avg=self.total/3

        if self.name is None:
            print("Hello default student")
        else:
            print("Hello", self.name)

        print("Your average quiz score is", avg)


s1 = Student()
s1.quizcalc(10)
print('------------------------')
s1.printdetail()
s2 = Student("Harry")
s2.quizcalc(10, 8)
print('------------------------')
s2.printdetail()
s3 = Student("Hermione")
s3.quizcalc(10, 9, 10)
print('------------------------')
s3.printdetail()
```

```python
#Task 6
class Vehicle:
    def __init__(self):
        self.x=0
        self.y=0

    def moveUp(self):
        self.y=self.y+1

    def moveLeft(self):
        self.x=self.x-1

    def moveDown(self):
        self.y=self.y-1

    def moveRight(self):
        self.x=self.x+1

    def print_position(self):
        print("("+str(self.x)+","+str(self.y)+")")


car=Vehicle()
car.print_position()
car.moveUp()
car.print_position()
car.moveLeft()
car.print_position()
car.moveDown()
car.print_position()
car.moveRight()
```

```python
#Task 9
class Patient:

    def __init__(self,name,age):
        self.name=name
        self.age=age
        self.symptoms=[]

    def add_Symptom(self,*args):
        for i in args:
            self.symptoms.append(i)

    def printPatientDetail(self):
        print("Name: {}".format(self.name))
        print("Age: {}".format(self.age))
        print("Symptoms: ",end="")
        print(", ".join(self.symptoms))


p1 = Patient("Thomas", 23)
p1.add_Symptom("Headache")
p2 = Patient("Carol", 20)
p2.add_Symptom("Vomiting", "Coughing")
p3 = Patient("Mike", 25)
p3.add_Symptom("Fever", "Headache", "Coughing")
print("=========================")
p1.printPatientDetail()
print("=========================")
p2.printPatientDetail()
print("=========================")
p3.printPatientDetail()
```

```python
#Task 11
class Shinobi:

    def __init__(self,name,rank):
        self.name=name
        self.rank=rank
        self.salary=0
        self.mission=0

    def changeRank(self,new_rank):
        self.rank=new_rank

    def calSalary(self,new_mission):
        self.mission=new_mission

        if self.rank=='Genin':
            self.salary=self.mission * 50
        elif self.rank=='Chunin':
            self.salary=self.mission * 100
        else:
            self.salary=self.mission * 500

    def printInfo(self):
        print('Name: ',self.name)
        print('Rank: ',self.rank)
        print('Number of mission:',self.mission)
        print('Salary: ',self.salary)


if __name__ == '__main__':
    naruto = Shinobi('Naruto', 'Genin')
    naruto.calSalary(5)
    naruto.printInfo()
    print('=========================')
    shikamaru = Shinobi('Shikamaru', 'Genin')
    shikamaru.printInfo()
    shikamaru.changeRank('Chunin')
    shikamaru.calSalary(10)
    shikamaru.printInfo()
    print('=========================')
    neiji = Shinobi('Neiji', 'Jonin')
    neiji.calSalary(5)
    neiji.printInfo()
```

```python
#Task 13
class Batsman:
    ballsFaced=0
    runsScored=0
    name="New Batsman"

    def __init__(self,*args):
        if len(args)>2:
            self.name=args[0]
            self.runsScored=args[1]
            self.ballsFaced=args[2]
        else:
            self.runsScored=args[0]
            self.ballsFaced=args[1]

    def setName(self,name):
        self.name=name

    def printCareerStatistics(self):
        print("Name: ",self.name)
        print("Runs Scored: {} ,Balls Faced: {}".format(self.runsScored, self.ballsFaced))

    def battingStrikeRate(self):
        st_rate=(self.runsScored/self.ballsFaced)*100
        return st_rate


b1 = Batsman(6101, 7380)
b1.printCareerStatistics()
print("================================")
b2 = Batsman("Liton Das", 678, 773)
b2.printCareerStatistics()
print("--------------------------------")
print(b2.battingStrikeRate())
print("================================")
b1.setName("Shakib Al Hassan")
b1.printCareerStatistics()
print("--------------------------------")
print(b1.battingStrikeRate())
```

```python
#Task 14
class EPL_Team:
    no_of_title=0

    def __init__(self,name,song=None):
        self.name = name
        self.song = song

        if song==None:
            self.song="No Slogan"

    def increaseTitle(self):
        self.no_of_title+=1

    def changeSong(self,new_song):
        self.song=new_song

    def ShowClubInfo(self):
        return "Name : {}\nSong: {}\nTotal No. of title: {}".format(self.name, self.song, self.no_of_title)


manu = EPL_Team("Manchester United", "Glory Glory Man United")
chelsea = EPL_Team("Chelsea")
print("========================================")
print(manu.ShowClubInfo())
print("########################################")
manu.increaseTitle()
print(manu.ShowClubInfo())
print("========================================")
print(chelsea.ShowClubInfo())
chelsea.changeSong("Keep the Blue Flag flying high")
print(chelsea.ShowClubInfo())
```

```python
#Task 15
class Account():

    def __init__(self, name='Default Account',balance=0.0):
        self.name=name
        self.balance=balance

    def details(self):
        return f'{self.name}\n{self.balance}'

    def withdraw(self, amount):
        if amount>self.balance:
            print("Sorry, Withdraw unsuccessful!The account")
            print("balance after deducting withdraw amount is")
            print("equal to or less than minimum.")
        else:
            self.balance-=amount
            print('Withdraw successful! New balance is', self.balance)


a1 = Account()
print(a1.details())
print('----------------------')
a1.name = 'Oliver'
a1.balance = 10000.0
print(a1.details())
print('----------------------')
a2 = Account('Liam')
print(a2.details())
print('----------------------')
a3 = Account('Noah', 400)
print(a3.details())
print('----------------------')
a2.withdraw(6930)
print('----------------------')
a2.withdraw(600)
print('----------------------')
a1.withdraw(6929)
```

```python
#Task 16
class Author:

    def __init__(self, name="Default",*books):
        self.name=name
        self.books=[]
        for book in books:
            self.books.append(book)

    def addBooks(self,*books):
        for book in books:
            self.books.append(book)

    def changeName(self, name):
        self.name=name

    def printDetails(self):
        print("Author Name:",self.name)
        print("-------")
        print("List of Books:")
        
        for book in self.books:
            print(book)


auth1 = Author("Humayun Ahmed")
auth1.addBooks("Deyal", "Megher Opor Bari")
auth1.printDetails()
print("======================")
auth2 = Author()
print(auth2.name)
auth2.changeName("Mario Puzo")
auth2.addBooks("The Godfather", "Omerta", "The Sicilian")
print("======================")
auth2.printDetails()
print("======================")
auth3 = Author("Paulo Coelho", "The Alchemist", "The Fifth Mountain")
auth3.printDetails()
```
